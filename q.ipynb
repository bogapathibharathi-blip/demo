```python
# 1 to 20
#jackpot number =15
#10
# choose the higher number
# 19
# choose the lower number
#15
# your attempt in 3 times
import random 
jackpot=random.randint(1,20)
num=int(input('enter a number 1-20'))
c=1
while jackpot!=num:
    if num<jackpot:
        print("choose the higher value")
    else:
        print("choose the lower value")
    num=int(input('enter a number 1-20'))
    c+=1
else:
    print(f' you won the game {num}')
    print(" your attempt",c)
```
